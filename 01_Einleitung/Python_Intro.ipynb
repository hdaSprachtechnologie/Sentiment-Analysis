## 6. Funktionsdefinitionen

Manchmal möchte man, dass der Code nicht direkt ausgeführt wird, sondern dass eine Funktion definiert wird, die man wiederverwenden kann.
Der Code wird nicht direkt ausgeführt, sondern erst, wenn die Funktion aufgerufen wird.


```python
# Hier z.B. unser Code für die Fibonacchi-Zahlen, wobei man angeben kann, bis zu welcher Zahl der Code ausgeführt werden soll.

def fibo(n):
    a, b = 0, 1
    while b < n:
        print(b)
        a, b = b, a+b
```

```python
# Probieren Sie es doch mal!

fibo(15)
```

Eine Funktion kann anstelle eines "Print"-Befehls für die Ausgabe einen "Return"-Wert haben, sodass der Ausgabewert später wiederverwendet werden kann. 

```python
def fibo_return(n):
    result = []
    a, b = 0, 1
    while b < n:
        result.append(a)
        a, b = b, a+b
    return result
```

```python
fibo_return(10)
```

## 8. Dateien einlesen und in Dateien schreiben

In manchen Fällen möchte man den Text aus einer Datei einlesen.

Als Beispiel nehmen wir den Text aus "Werther" von Goethe, den Sie als Textdatei in Moodle finden.


```python
# Mit "open" öffnet man eine Datei. Die Option "r" bedeutet, dass die Datei zum Lesen (read) geöffnet wird. Unter "encoding" steht die Kodierung der Datei.

textin = open('goethes_werther.txt', 'r', encoding='utf-8')

# Der Text wird zeilenweise eingelesen. 

lines = textin.readlines()

# Anschließend kann man die Zeilen einzeln verarbeiten.

for line in lines:
    if line.startswith("W"):
        print(line)

# Wir müssen aber feststellen, dass der Werther-Text nur eine Zeile enthält. Mit welchem Buchstaben beginnt diese?

# Dateien müssen wieder geschlossen werden.

textin.close()
```

```python
# Mit der Option "w" öffnen wir eine Datei zum Schreiben (write)

out = open('out.txt','w',encoding='utf-8')

# Nun schreiben wir etwas hinein

out.write('Hallo')

# Dateien müssen wieder geschlossen werden.

out.close()
```

Hier sortieren wir die Wörter im Satz nach Wortlänge.

```python
# Einen String definieren und mit split in eine Wortliste verwandeln.
words = 'Die Katze fängt die Maus'.split()
print(words)

# Ein Tuple für jedes Wort erzeugen, das die Wortlänge und das Wort enthält.
wordlens = [(len(word),word) for word in words]
print(wordlens)

# Diese Liste von Tuples sortieren
wordlens.sort()
print(wordlens)

# Wieder einen String aus den Wörtern machen, dabei die Wortlänge ignorieren
'  '.join(w for(_,w) in wordlens)
```

Die Unterschiede von Listen und Tuples: 

| Tuples | Listen   |

| festgelegte Länge | unterschiedliche Länge |
| Reihenfolge ist relevant | Reihenfolge ist nicht relevant |

### Dictionaries

Dictionaries heißen in anderen Programmiersprachen "Arrays". Man kann damit Einträgen Werte zuweisen. Beispiele sind: Wörter und ihre grammatischen Kategorien, Wörter und ihre Grundformen, Wörter und ihre Übersetzungen, Namen und Telefonnummern, Wörter und ihre Sentiment-Polarität und vieles mehr.

Nehmen wir als Beispiel den Satz "Farblose grüne Ideen schlafen wild" (im Englischen "colourless green ideas sleep furiously"), der von Noam Chomsky als Beispiel für einen syntaktisch korrekten aber semantisch falschen Satz erfunden wurde.

In unserem Dictionary ordnen wir jedem Wort eine grammatische Kategorie (POS) zu.

```python
# Dictionary öffnen

pos = {}

# Wörtern ihre POS zuordnen. Man sagt: Der Key "farblose" bekommt den Wert "ADJ"

pos['farblose'] = 'ADJ'
pos['grüne'] = 'ADJ'
pos['Ideen'] = 'N'
pos['schlafen'] = 'V'
pos['wild'] = 'ADV'

pos
```

```python
# Das geht übrigens auch kompakter

pos = {'Ideen':'N','farblose':'ADJ','grüne':'ADJ','schlafen':'V','wild':'ADV'}
```

```python
# Auf Werte zugreifen

pos['Ideen']
```

```python
# Eine Liste der Wörter

list(pos)
```

```python
# Sortieren

sorted(pos)
```

```python
# Wörter, die mit 'n' enden

[w for w in pos if w.endswith('n')]
```

```python
# Alles als Tabelle aufführen

for word in sorted(pos):
    print(word + ":", pos[word])
```

```python
# Aber was machen wir mit der Katze?

pos['Katze']
```

```python
# Wir definieren einen "Default-Wert": Alle unbekannten Wörter sollen Nomen sein

from collections import defaultdict

pos = defaultdict(lambda: 'N')

pos['Katze']
```
